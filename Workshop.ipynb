<font color=#505050 size=3 face='arial black'>From the point of view of the data scientist, Python is a tool for collecting and manipulating data. Most data is in the form of text and numbers. So Python has a simple system for representing numbers and text.
<br><br>
There are four basic types of data in Python: <font color=#13577F>boolean</font>, <font color=#13577F>strings</font>, <font color=#13577F>floats</font> and <font color=#13577F>integers</font>. Strings are text. Integers are integers, one of the two types of numbers. Floats are the equivalent of decimals. Booleans represent the truth value of various statements that we might want to ask Python.</font>

```python
print type(24) #integer
print type(2.4) #float
print type("It was the White Rabbit") #string
print type(True) #boolean
print type(min) #function
```

```python
original_sentence = "Down, down, down. Would the fall NEVER come to an end!"
print "original       :", original_sentence

no_commas = original_sentence.replace(",","")
print "no commas      :", no_commas

no_exclamations = no_commas.replace("!","")
print "no exclamations:", no_exclamations

no_stops = no_exclamations.replace(".","")
print "no exclamations:", no_exclamations

split_sentence = no_stops.split(" ")
print "list           :", split_sentence
```

```python
original_sentence.replace(",","").replace("!","").replace(".","")
```

```python
"".join([letter for letter in original_sentence if letter not in string.punctuation])
```

```python
import re
re.sub('[!,.]','',original_sentence)
```

```python
file = open("AliceInWonderland.txt")
text = file.read()
text = text.split()
data = map(len, text)
plt.hist(data,color="#ff6347")
plt.xlim(0,16)
plt.show()
```

```python
np.linspace(0.5,16.5,17)
```

```python
plt.hist(data,bins=np.linspace(0.5,16.5,17),color="#ff6347")
plt.xlim(0,16)
plt.title('Distribution of word lengths in P & P')
plt.xlabel('word length')
plt.ylabel('frequency')
plt.show()
```
